#step3

<div id="section1"   dir='rtl'>
    <h2>در این گام از ما خواسته شده 10 نمادی که در هرماه بیشترین میزان افزایش قیمت را داشتند را پیدا کنیم، برای اینکار از  window و row_number استفاده می کنیم به اینصورت که ابتداد براسای نماد و ماه گروه بندی می کنیم و بر روی ستوت close_price_change ماکسمیمم می گیریم چون این ستون اختلاف دو مقدار پایانی و تغییر است     </div>
</h2>
        <hr>

````python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

df_maxprice = cases.groupBy('symbol', 'month').max('close_price_change')
````

<div id="section1"   dir='rtl'>
    <h2>سپس یک ستون جدید بر روی جدول قبل به اسم رنک ایجاد می کنیم  وبر روی ماه بر اساس تابع پارتیشن در window، پارتیشن می کنیم و بر اساس ماکسمیمم اختلاف در هر ماه سورت نزولی می کنیم تا بیشترین قیمت ها به ترتیب در بالا قرار گیرند سپس با استفاده از تابع فیلتر، 10 تا ماکسیمم هر ماه که ستون رنکشان کمتر از 10 را نمایش می دهیم  </div>
</h2>
        <hr>

````python
df_maxprice=df_maxprice.withColumn('rank',row_number().over(Window.partitionBy('month').orderBy(col("max(close_price_change)").desc()))).filter(col('rank')<=10)
df_maxprice
````

<div id="section1"   dir='rtl'>
    <h2>همانطور که می بینیم تعداد سطرهای این جدول 70 است که شامل 7 ماه 10 تایی سمبل است</div>
</h2>
        <hr>

````python
df_maxprice.count()
````

    <h2>و در نهایت بر اساسماه و رنک سورت صعودی می کنیم </div>
</h2>
        <hr>

````python
df_maxprice.orderBy(col("month").asc(),col("rank").asc())
````

<div id="section1"   dir='rtl'>
    <h2>برای sql چندین گام را پیش می بریم به این صورت که ابتدا نماد و ماه و ماکسمیمم ستون close_price_change را انتخاب می کنیم به طوری که بر اساس ماه و نماد گروه بندی می کنیم</div>
</h2>
        <hr>

sql

````python
cases.registerTempTable('Stock_Exchange_Daily')
sqlDF_maxprice=spark.sql("""select symbol,month,max(close_price_change) as max_close
FROM Stock_Exchange_Daily  GROUP BY symbol,month """)

sqlDF_maxprice
 
````

<div id="section1"   dir='rtl'>
    <h2>سپس با استفاده پارتیشن بندی بر روی ماه و سورت نزولی براساس ماکسیمم یک ستون جدید که رنک است به این جدول اضافه می کنیم</div>
</h2>
        <hr>

````python
sqlDF_maxprice.registerTempTable('sqlDF_maxprice')
sql1 = spark.sql("""SELECT * , ROW_NUMBER() OVER(PARTITION BY month ORDER BY max_close DESC) as rn FROM sqlDF_maxprice""")
sql1
````

<div id="section1"   dir='rtl'>
    <h2>در نهایت با یک شرط بر روی ستون رنک(rn) تمام سطر هایی را انتخاب می کنیم که رنکشان کمتر از 10 هستند و براساس ماه و رنک صورت صعودی می کنیم</div>
</h2>
        <hr>

````python
sql1.registerTempTable('sql1')

spark.sql("""SELECT * FROM sql1 where rn<=10 ORDER BY month ASC,rn ASC""")

````
