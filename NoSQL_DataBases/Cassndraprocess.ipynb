```python
import pandas as pd
import numpy as np
```

```python
data4= pd.read_csv('fma_dataset (1).csv')
data4
```

```python
print (data4['date_released'].dtype)
```

```python
data4.isnull().sum()
```

# split date_released(timespan) column to two columns (date and time)

```python
data4[["date", "time"]] = data4['date_released'].apply(lambda x: pd.Series(str(x).split(" ")))
```

```python
data4
```

# split date column to year,month and day

```python
data4[["year", "month", "day"]] = data4["date"].str.split("-", expand=True)
```

```python
data4
```

```python
print (data4['month'].dtype)
```

# process specific character

### first: processing character such as  ; ..we replace them with nothing and  we can find nan value in columns so we should replace nan value with forinstence 0 for processing

```python
#spec_chars = {'|':'','\"':'',"$":'','\'':'',';':'',"#":''}
#for char in spec_chars:
#    data4.replace(char, regex=True)
#data4= data4.replace('nan', 0,regex=True)
```

```python
for i in data4.columns:
    display(data4[data4[i]=='nan'])
```

```python
data4= data4.replace({'\"':'','&':'',';':'','`':'','#':'','%':'','\'':'',';':'',
                     '^':'','|':'','nan':0}, regex=True)
```

```python
# for \ and | we should use regex
data4 = data4.replace(to_replace= r'\\', value= '', regex=True)
data4= data4.replace(to_replace= r'\|', value= '', regex=True)
```

### second: title track of one row was empty and it make problem to copy in cassandra so I do below proccess

```python
data4.iloc[290:300]
```

```python
data4=data4.replace({"":'Null'})
#data4=data4.replace.replace({"\"\"":'Null'})
```

```python
data4.iloc[290:300]
```

# process missing value 

```python
data4[['duration', 'favorites_artist','listens_track','favorites_track','listens_album','favorites_album','year','month','day']] = data4[['duration', 'favorites_artist','listens_track','favorites_track','listens_album','favorites_album','year','month','day']].fillna(value=0)
```

# delete zero (forinstance format 05) from day and month and convert to just formt 5

```python
data4['month']= data4['month'].astype('int64')
data4['day']= data4['day'].astype('int64')
```

```python
data4
```

```python
data4[['title_track', 'date_released','title_album','artist','genre','time']] = data4[['title_track', 'date_released','title_album','artist','genre','time']].fillna(value='Null')
```
